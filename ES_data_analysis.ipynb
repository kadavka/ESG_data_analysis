```python
import numpy as np
import pandas as pd
from scipy import stats

import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
#S&P 500 companies and their ESG risk scores

data = pd.read_csv('/Users/marekkadavy/Desktop/pythonProjects/ESG_scores/SP_500_ESG_Risk_Ratings.csv')
data.head()
```

```python
data.describe()
```

```python
def check(data):
    l = []
    columns = data.columns
    for col in columns:
        instances = data[col].count() #number of null entries in the current column
        dtypes = data[col].dtype  #data type of the current column
        unique = data[col].nunique() #number of unique values in the current column
        sum_null = data[col].isnull().sum() #number of NA/null values in the current column
        duplicates = data.duplicated().sum() # number of duplicated rows in the entire dataframe
        l.append([col, dtypes, instances, unique, sum_null, duplicates])
    data_check = pd.DataFrame(l, columns=["column", "dtype", "instances", "unique", "sum_null", "duplicates"])
    return data_check

data_summary = check(data)
print(data_summary)
```

```python
data = data.dropna(subset=['Sector','Total ESG Risk score'])

data = data.reset_index(drop=True) # reset indexing after removing 0 values
```

```python
#How are different SECTORS represented
sector_representation = data['Sector'].value_counts().sort_values(ascending=False)

#Plot style
top_n = 10
sector_representation = data['Industry'].value_counts().nlargest(top_n)
cmap = plt.cm.get_cmap('PRGn')  # 'PRGn' is a colormap with purple to yellow gradient
colors = [cmap(i) for i in np.linspace(0, 1, top_n)]

plt.figure(figsize=(10, 8))
sector_representation.plot(kind='bar', color=colors)
plt.title('Representation of Different Sectors')
plt.ylabel('Number of Occurrences')
plt.xlabel('Sector')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#How are different INDUSTRIES represented
#Top 20 Industries represented in the dataset

industry_representation = data['Industry'].value_counts().sort_values(ascending=False)

top_n = 20  # Select top 20 industries to display
industry_representation = data['Industry'].value_counts().nlargest(top_n)


plt.figure(figsize=(10, 8))
industry_representation.plot(kind='bar', color=colors)
plt.title('Representation of Different Industries')
plt.ylabel('Number of Occurrences')
plt.xlabel('Industry')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Plot style
industry_representation = data['Industry'].value_counts().nlargest(top_n)
cmap = plt.cm.get_cmap('PRGn')  # 'PRGn' is a colormap with purple to yellow gradient
colors = [cmap(i) for i in np.linspace(0, 1, top_n)]
```

```python
#How are different and sectors represented in the dataset

#Sector-wise risk

sector_risk = data.groupby('Sector')['Total ESG Risk score'].mean().sort_values()

plt.figure(figsize=(10, 8))
sector_risk.plot(kind='barh')
plt.title('Average Total ESG Risk Score by Sector')
plt.ylabel('Sector')
plt.xlabel('Average Total ESG Risk Score')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

#add annotation
for i, v in enumerate(sector_risk):
    plt.text(v + 0.2, i, str(f"{v:.2f}"), color='black', verticalalignment='center')

plt.show()
```

```python
#Companies with the top ESC performance (=lowest ESG score)

top_score = data.sort_values(by='Total ESG Risk score', ascending=True).head(10) # 10 best companies



plt.figure(figsize=(10, 6))  
top_score.plot(x='Name', y='Total ESG Risk score', kind='barh', legend=False, color='green')
plt.title('Top 10 Performers by ESG Risk Level')
plt.ylabel('Name') 
plt.xlabel('ESG Risk Score')
plt.xticks(rotation=45, ha='right')
plt.gca().invert_yaxis()
plt.tight_layout()

#add annotation
for i, v in enumerate(top_score['Total ESG Risk score']):
    plt.text(v + 0.2, i, str(v), color='black', verticalalignment='center')

plt.show()
```
